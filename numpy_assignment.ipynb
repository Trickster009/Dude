# Task 6

```python
sales = np.array([1200,1500,900,2000,1800,1700,1600])
sales
```

```python
print("Total weekly sales:" ,np.sum(sales))
```

```python
print("Average daily sales:", np.average(sales))
```

```python
print("Highest day sales:", np.max(sales))
print("Lowest day sales:", np.min(sales))
```

```python
print("Standard deviation of sales:", np.std(sales))
```

```python
above_avg_sales = []
avgr=np.average(sales)
for i in sales:
    if i > avgr:
        above_avg_sales.append(i)
above_avg_sales
```
