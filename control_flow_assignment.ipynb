```python
order_amount = input("Enter order amount:")

if order_amount.isdigit():
    order_amount=int(order_amount)

    if order_amount >= 2000:
        discount = 0.15
    elif order_amount in range(1500, 2000):
        discount = 0.1
    elif order_amount in range(1000, 1500):
        discount = 0.07
    else:
        discount = 0
    
    final_amount = order_amount - (order_amount * discount)
    print(final_amount)

else:
    print("Enter a number")
```

# Task 2

```python
orders = [1200,2500,800,1750,3000]
Total_Rev_Post_Discount = 0
for order_amount in orders:
    if order_amount >= 2000:
        discount = 0.15
        discount_percentage = "15%"
    elif order_amount in range(1500, 2000):
        discount = 0.1
        discount_percentage = "10%"

    elif order_amount in range(1000, 1500):
        discount = 0.07
        discount_percentage = "7%"
    else:
        discount = 0
        discount_percentage = "0%"
    
    final_amount = order_amount - (order_amount * discount)
    print(order_amount,discount_percentage,final_amount)
    Total_Rev_Post_Discount += final_amount
print("Total revenue post discount: " ,Total_Rev_Post_Discount)
```

# Task 3

```python

orders = []
final_amount_list = []
Total_Rev_Post_Discount = 0
while(True):
    print("1--Add order amount to the running list")
    print("2--Show all orders and totals after applying discounts")
    print("q--Quit") 

    option = input("Choose any of the given option")
    
    if option == "1":
        print("Enter an order amount:")
        order_amount = int(input())
        orders.append(order_amount)

    
    elif option == "2":
        for order_amount in orders:
            if order_amount >= 2000:
                discount = 0.15
            elif order_amount in range(1500, 2000):
                discount = 0.1
            elif order_amount in range(1000, 1500):
                discount = 0.07
            else:
                discount = 0

                
            final_amount = order_amount - (order_amount * discount)
            final_amount_list.append(final_amount)
            Total_Rev_Post_Discount += final_amount 
        print("Orders:", orders)
        print("Final amounts after discount:",final_amount_list)
        print("Total order amount post discount: " ,Total_Rev_Post_Discount)

    elif option == "q":
        print("Exiting program...")
        break

    else:   
        print("Select any of the above 3 options only")
        continue
```

# Task 5 

```python
daily = [200,150,0,400,50,-1,300]
total_sales = 0
ctr = 0
for i in daily:
    if i == -1:
        print("Data corrupted, exiting program")
        break
    elif i == 0:
        print("No sales continuing further")
        continue
    elif i > 0 :
        total_sales += i
        ctr += 1
        print(f"{ctr}.Sale = {total_sales}")
print("Total_sales: ",total_sales)
```
